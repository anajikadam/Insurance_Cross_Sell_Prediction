# Insurance Cross Sell Prediction using Machine Learning and Deep Learning

## Problem Statement:

Our client, an insurance company, needs help in building a model to predict the likelihood of past health insurance policyholders showing interest in purchasing vehicle insurance from the company. The company charges premiums from customers to provide them with compensation in case of specified loss, damage, illness, or death. By predicting the interest in vehicle insurance, the company can plan its communication strategy and optimize its revenue. The model will be built using information about demographics, vehicles, and insurance policies.

- Driving_License: Customer have DL (0/1)
- Region_code: code for region to identify location
- Previously_Insured: customer has previous vehicle insurance (0/1)
- Vehicle_age: Vehicle age
- Vehicle_Damage: previously vehicle damage or not (Yes/No)
- Annual Premium: cost of Annual primium insurance
- policy_salse_channel: communication medium to connect customers( annonyms code)
- Vintage: how much time (number of days) customers with the company
- Response: customer is intersted in take insurance from company (Target column)(1/0)

```python
X = df.drop('Response', axis = 1)
y = df['Response']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# SMOTE works by utilizing a k-nearest neighbour algorithm to create synthetic data. 
# SMOTE first starts by choosing random data from the minority class, then k-nearest neighbors from the data are set. 
# Synthetic data would then be made between the random data and the randomly selected k-nearest neighbor.

oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
folds = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 40)
```

```python
def grid_search(model,folds,params,scoring):
    grid_search = GridSearchCV(model,
                                cv=folds, 
                                param_grid=params, 
                                scoring=scoring, 
                                n_jobs=-1, verbose=1)
    return grid_search
```

```python
def print_best_score_params(model):
    print("Best Score: ", model.best_score_)
    print("Best Hyperparameters: ", model.best_params_)
```

```python
log_reg = LogisticRegression()
log_params = {'C': [0.01, 1, 10], 
          'penalty': ['l1', 'l2'],
          'solver': ['liblinear','newton-cg','saga']
         }

         
grid_search_log = grid_search(log_reg, folds, log_params, scoring=None)
grid_search_log.fit(X_train, y_train)
print_best_score_params(grid_search_log)
```

```python
dtc = DecisionTreeClassifier(random_state=40)
dtc_params = {
    'max_depth': [5,10,20,30],
    'min_samples_leaf': [5,10,20,30]
}
grid_search_dtc = grid_search(dtc, folds, dtc_params, scoring='roc_auc_ovr')
grid_search_dtc.fit(X_train, y_train)
print_best_score_params(grid_search_dtc)
```
